# Purpose of this note book is to clean and merged all the data found under tweets/filtered_tweets folder

```python
import pandas as pd
import numpy as np

import os
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
merged_df = None

directory = '../datasets/tweets/filtered_tweets'
for filename in os.listdir(directory):
    file_path = os.path.join(directory, filename)

    if os.path.isfile(file_path):
        if (filename[-12:] == "filtered.csv"):
            print(filename)
            df = pd.read_csv(file_path)
            df = df.loc[:, ~df.columns.str.contains('^Unnamed')] #dropped unneeded columns
            if 'username' in df.columns:
                df.rename(columns={'username':'user_name'}, inplace=True)
            if (merged_df is None):
                merged_df = df
            else:
                merged_df = pd.concat([merged_df, df], ignore_index=True, sort=False)
                merged_df.reset_index(drop=True, inplace=True)
            print(df.shape)
            print(merged_df.shape)
```

```python
#fill na values with reference to existing data
merged_df['retweets'] = merged_df['retweets'].fillna(merged_df.groupby('user_name')['retweets'].transform(
    lambda x: x.fillna(x.mean())
))

merged_df['favorites'] = merged_df['favorites'].fillna(merged_df.groupby('user_name')['favorites'].transform(
    lambda x: x.fillna(x.mean())
))

merged_df['isRetweet'] = merged_df['isRetweet'].fillna(merged_df.groupby('user_name')['isRetweet'].transform(
    lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x
))

merged_df['isFlagged'] = merged_df['isFlagged'].fillna(merged_df.groupby('user_name')['isFlagged'].transform(
    lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x
))
```

```python
# fill with dummy values, since no existing data
merged_df['retweets'] = merged_df['retweets'].fillna(-1)
merged_df['isRetweet'] = merged_df['isRetweet'].fillna('False')
merged_df['favorites'] = merged_df['favorites'].fillna(-1)
merged_df['isFlagged'] = merged_df['isFlagged'].fillna('f')
```
