```python
import pandas as pd
import numpy as np

import os
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
chains = ['bit', 'doge', 'eth', 'polygon', 'solana']
types = ['gas', 'transac', 'avg']


directory = './from_bigquery'

for chain in chains:
    for typ in types:
        merged_df = None
        for filename in os.listdir(directory):
            file_path = os.path.join(directory, filename)
            
            if os.path.isfile(file_path):
                
                if (f"{chain} {typ}" in filename):
                    print(filename)
                    df = pd.read_csv(file_path)
                    if (merged_df is None):
                        merged_df = df
                    else:
                        merged_df = pd.concat([merged_df, df], ignore_index=True, sort=False)
                        merged_df.reset_index(drop=True, inplace=True)
                    print(df.shape)
                    print(merged_df.shape)
        if merged_df is not None:
            merged_df.to_csv(f"{chain}_{typ}_all.csv")
        
```
